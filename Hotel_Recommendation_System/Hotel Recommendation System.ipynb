#### As 440 rows has "none" gender, so filling them using my gender prediction model that I have made using XgBoost

````python
# importing the predicted genders dataframe
predicted_gender = pd.read_csv("unknown_gender_with_predictions.csv")
````

````python
predicted_gender.head(3)
````

````python
# merging the user dataframe with the precited gender dataframe on the basis of code
user_merged_df = pd.merge(users_df, predicted_gender, how ="left", on = "code")
user_merged_df.head()
````

````python
# replacing none values in gender with the predicted gender
user_merged_df['gender'] = user_merged_df['gender'].replace('none', pd.NA)
user_merged_df['gender'] = user_merged_df['gender'].fillna(user_merged_df['predicted_gender'])
````

````python
# checking for the none value in the gender column
user_merged_df[user_merged_df['gender'] == "none"]
````

````python
# merging the user dataframe with the hotel dataframe
main_df = pd.merge(user_merged_df, hotels_df, how = "inner", left_on = "code", right_on = "userCode")
````

````python
# viewing the main dataframe
main_df.head()
````

# My Hotel Recommendation Model



````python
def recommendation_system(place, price_range, days_range):
    filtered_hotels_by_place = hotel_features[hotel_features['place'] == place]
    filtered_hotels_by_price = filtered_hotels_by_place [(filtered_hotels_by_place ['price'] >= price_range[0]) & (filtered_hotels_by_place ['price'] <= price_range[1])]
    filtered_hotels_by_day = filtered_hotels_by_price[(filtered_hotels_by_price['days'] >= days_range[0]) & (filtered_hotels_by_price['days'] <= days_range[1])]
    return filtered_hotels_by_day.head()

recommendation_system('Recife (PE)', price_range=(300, 350), days_range=(3, 4))
````

# Explanation 

In this code, I have built a hotel recommendation system that recommends hotels based on three user inputs:

- Place: The location where the user wants to book a hotel (e.g., "Recife (PE)").
- Price Range: A budget range for the price of hotels (e.g., 300 to 350).
- Days Range: The number of days the user plans to stay (e.g., 3 to 4 days).

The model filters the hotels step by step based on these inputs and provides the best matches.

**filtered_hotels_by_place = hotel_features[hotel_features['place'] == place]**
- This line filters the dataset (hotel_features) to include only the hotels located in the place specified by the user.
- For example, if the user selects Recife (PE), only the hotels in Recife will remain in the filtered dataset.

**filtered_hotels_by_price = filtered_hotels_by_place[
    (filtered_hotels_by_place['price'] >= price_range[0]) & 
    (filtered_hotels_by_place['price'] <= price_range[1])
]**
- Here, we further narrow down the dataset by selecting only those hotels whose price falls within the price range specified by the user.
- For example, if the user selects a price range of (300, 350), only hotels with prices between 300 and 350 will remain in the filtered dataset.

**filtered_hotels_by_day = filtered_hotels_by_price[
    (filtered_hotels_by_price['days'] >= days_range[0]) & 
    (filtered_hotels_by_price['days'] <= days_range[1])
]**
- Finally, this step filters the dataset further by selecting hotels where the number of days (days) falls within the days range provided by the user.
- For example, if the user specifies a stay duration of 3 to 4 days, only hotels offering this duration will remain.

After applying all filters, this line returns the top 5 hotels (.head()) from the filtered dataset that match all the criteria.
These are the hotels recommended to the user.

- Here’s how the function is used:

**recommendation_system('Recife (PE)', price_range=(300, 350), days_range=(3, 4))**
- place='Recife (PE)': The user wants to find hotels in Recife.
- price_range=(300, 350): The user has a budget of 300 to 350 for hotel prices.
- days_range=(3, 4): The user plans to stay for 3 to 4 days.

When executed, the function will:

- Filter hotels in Recife.
- Narrow the list to hotels priced between 300 and 350.
- Further filter hotels offering stays of 3 to 4 days.
- Return the top 5 matching hotels.
